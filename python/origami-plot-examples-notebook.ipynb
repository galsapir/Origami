```python
import pandas as pd
import matplotlib.pyplot as plt
from origami_plot import create_origami_plot, VariableConfig, ColorScheme

# Ensure inline plotting
%matplotlib inline
```

```python
# Example 1: Basic origami plot
example_data_1 = pd.Series({
    'Variable A': 50,
    'Variable B': 70,
    'Variable C': 30,
    'Variable D': 90
})

variable_configs_1 = [
    VariableConfig('Variable A', 0, 100),
    VariableConfig('Variable B', 0, 100),
    VariableConfig('Variable C', 0, 100),
    VariableConfig('Variable D', 0, 100)
]

fig, ax = create_origami_plot(example_data_1, variable_configs_1, ColorScheme.DEFAULT, 
                              title="Example 1: Basic Origami Plot")
plt.show()
```

```python
# Example 2: Origami plot with different range
example_data_2 = pd.Series({
    'Metric 1': 150,
    'Metric 2': 120,
    'Metric 3': 80,
    'Metric 4': 60
})

variable_configs_2 = [
    VariableConfig('Metric 1', 0, 200),
    VariableConfig('Metric 2', 0, 150),
    VariableConfig('Metric 3', 0, 100),
    VariableConfig('Metric 4', 0, 100)
]

fig, ax = create_origami_plot(example_data_2, variable_configs_2, ColorScheme.COLORBLIND_FRIENDLY, 
                              title="Example 2: Origami Plot with Different Ranges")
plt.show()
```

```python
# Example 3: Origami plot with more variables
example_data_3 = pd.Series({
    'Feature 1': 0.8,
    'Feature 2': 0.6,
    'Feature 3': 0.9,
    'Feature 4': 0.7,
    'Feature 5': 0.5,
    'Feature 6': 0.3
})

variable_configs_3 = [
    VariableConfig('Feature 1', 0, 1),
    VariableConfig('Feature 2', 0, 1),
    VariableConfig('Feature 3', 0, 1),
    VariableConfig('Feature 4', 0, 1),
    VariableConfig('Feature 5', 0, 1),
    VariableConfig('Feature 6', 0, 1)
]

fig, ax = create_origami_plot(example_data_3, variable_configs_3, ColorScheme.DEFAULT, 
                              title="Example 3: Origami Plot with More Variables")
plt.show()
```

```python
# Example 4: Origami plot with custom figure size
example_data_4 = pd.Series({
    'Aspect 1': 300,
    'Aspect 2': 450,
    'Aspect 3': 200,
    'Aspect 4': 350,
    'Aspect 5': 400
})

variable_configs_4 = [
    VariableConfig('Aspect 1', 0, 500),
    VariableConfig('Aspect 2', 0, 500),
    VariableConfig('Aspect 3', 0, 500),
    VariableConfig('Aspect 4', 0, 500),
    VariableConfig('Aspect 5', 0, 500)
]

fig, ax = create_origami_plot(example_data_4, variable_configs_4, ColorScheme.COLORBLIND_FRIENDLY, 
                              title="Example 4: Custom Figure Size", figsize=(8, 8))
plt.show()
```
